```python
#Guess game first way
import random
number=random.randint(1,100)

try:
    guess=int(input("Enter the number"))
    counter=1

    while True:
        if number!=guess:
            if number<guess:
                print("Guess lower!")
            elif number>guess:
                print("Guess higher!")
            guess=int(input("Enter the number"))
            counter+=1
        else:
            print("you have entered right number")
            print("The number is", number)
            print("Attempts you have taken",counter)
            break

except ValueError as e:
    print("You have entered invalid value",e)
```
