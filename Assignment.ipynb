```python
#Question - Create a function to calculate age till now.
from datetime import date

def calculate_age(dob):  
    
    birth_year, birth_month, birth_day = dob
    today = date.today()

    age = today.year - birth_year

    
    if (today.month, today.day) < (birth_month, birth_day):
        age -= 1

    return age
```

```python
#Question - Create a function to check age eligibility for given customer based on DOB. Function will take two input DOB and ELIGIBILITY age.
from datetime import date

def check_eligibility(dob, eligibility_age):
    birth_year, birth_month, birth_day = dob
    today = date.today()

    age = today.year - birth_year
    if (today.month, today.day) < (birth_month, birth_day):
        age -= 1

    if age >= eligibility_age:
        return f"Eligible (Age: {age})"
    else:
        return f"Not Eligible (Age: {age})"
```

```python
#Question - Calculate total hrs based on input sign in and sign out time 
from datetime import datetime

def calculate_total_hours(sign_in, sign_out):

    time_format = "%H:%M"

    in_time = datetime.strptime(sign_in, time_format)
    out_time = datetime.strptime(sign_out, time_format)

    if out_time < in_time:
        return "Sign-out time cannot be before sign-in time"

    total_time = out_time - in_time
    total_hours = total_time.total_seconds() / 3600

    return total_hours
```
