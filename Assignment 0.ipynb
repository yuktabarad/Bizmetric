Comprehensive list:

```python
#Question 1 - Write a list comprehension to generate squares of numbers from 1 to 10.
squares=[i**2 for i in range(1,11)]
```

```python
#Question 2 -  Create a list of even numbers between 1 and 50 using list comprehension.
even=[i for i in range(1,51) if i%2==0]
even
```

```python
#Question 3 -  Convert all strings in a list to uppercase using list comprehension
names=["yukta","orange","dipti","parth","python"]
upper=[name.upper() for name in names]
upper
```

```python
#Question 4 -   Given a list of integers, create a new list that contains only the positive numbers.
numbers=[-10,12,9,-2,-6,-8,9,90]
positive=[num for num in numbers if num<0]
positive
```

```python
#Question 5 - Create a list of tuples (num, num^2) for numbers 1 to 5.
final=[(num, num**2) for num in range(1,6)]
final
```

```python
#Question 6 - Extract all vowels from a given string using list comprehension.
text="Yukta Rajesh Barad. I am student at MIT WPU university."
vowels=[i for i in text if i.lower() in 'aeiou']
vowels
```

```python
# Question 7 - Flatten a 2D list using list comprehension.

list2d = [[1,2],[3,4],[5,6]]
flat = [j for i in list2d for j in i]
print(flat)
```

```python
# Question 8 - Replace all negative numbers in a list with 0 using list comprehension.

nums = [-3,5,-1,8]
new = [0 if i<0 else i for i in nums]
print(new)
```

```python
# Question 9 - Given a list of words, create a list of lengths of each word.

words = ["yukta","data","analysis"]
lengths = [len(i) for i in words]
print(lengths)
```

```python
# Question 10 - Filter out words that start with the letter 'A' or 'a'.

words = ["Apple","banana","Ant","cat"]
new_words = [i for i in words if not i.lower().startswith("a")]
print(new_words)
```

```python
# Question 11 - From a list of numbers, generate a list of "even" or "odd" strings.

nums = [1,2,3,4,5,6]
result = ["even" if i%2==0 else "odd" for i in nums]
print(result)
```

```python
# Question 12 - Create a list of numbers divisible by both 3 and 5 in range 1–100.

list1 = [i for i in range(1,101) if i%3==0 and i%5==0]
print(list1)
```

```python
# Question 13 - Generate a multiplication table for 1–5 using nested list comprehension.

table = [[i*j for j in range(1,6)] for i in range(1,6)]
print(table)
```

```python
# Question 14 - Convert a dictionary’s keys into a list using list comprehension.

d = {"a":1,"b":2,"c":3}
key= [i for i in d]
print(key)
```

```python
# Question 15 - Extract numeric digits from a string using list comprehension.

text = "abc123xyz45"
digits = [i for i in text if i.isdigit()]
print(digits)
```

```python
# Question 16 - Remove all spaces from a string using list comprehension.

text = "Yukta Rajesh Barad"
new_text = [i for i in text if i!=" "]
print("".join(new_text))
```

```python
# Question 17: Create a list of characters that appear more than once in a string.

text = "programming"
result = [i for i in text if text.count(i)>1]
print(list(set(result)))

```

```python
# Question 18 - From a list of sentences, generate a list of all words.

sentences = ["My name is Yukta","I am learning python"]
words = [j for i in sentences for j in i.split()]
print(words)
```

```python
# Question 19 - Create a list of unique elements from a list.

nums = [1,2,2,3,4,4,5]
unique = [i for i in nums if nums.count(i)==1]
print(unique)
```

```python
# Question 20 - Generate all pairs (x, y) from list A and list B.

A = [1,2,3]
B = ["a","b"]
pairs = [(i,j) for i in A for j in B]
print(pairs)
```

Lambda function


```python
# Question 1: Write a lambda to add two numbers.

add = lambda a,b : a+b
print(add(5,3))
```

```python
# Question 2: Create a lambda to check if a number is even.
check = lambda x : x%2==0
print(check(4))
```

```python
# Question 3: Write a lambda to get the last character of a string.

last_char = lambda s : s[-1]
print(last_char("yukta"))
```

```python
# Question 4: Use lambda with map() to square every number in a list.
nums = [1,2,3,4]
result = list(map(lambda x:x*x, nums))
print(result)
```

```python
# Question 5: Use lambda with filter() to get only odd numbers from a list.
nums = [1,2,3,4,5]
result = list(filter(lambda x:x%2!=0, nums))
print(result)
```

```python
# Question 6: Use sorted() + lambda to sort a list of tuples by second value.
data = [(1,5),(2,1),(3,3)]
result = sorted(data, key=lambda x:x[1])
print(result)
```

```python
# Question 7: Create a lambda to check if a string is a palindrome.

pal = lambda s : s==s[::-1]
print(pal("madam"))
```

```python
# Question 8: Use lambda to find maximum of three numbers.
max3 = lambda a,b,c : max(a,b,c)
print(max3(3,7,5))
```

```python
# Question 9: Write a lambda to reverse a string.
rev = lambda s : s[::-1]
print(rev("python"))
```

```python
# Question 10: Use lambda with map() to convert a list of strings to integers.
nums = ["1","2","3"]
result = list(map(lambda x:int(x), nums))
print(result)

```

```python
# Question 11: Use lambda with filter() to remove empty strings from a list.
words = ["Yukta","","yamaha",""]
result = list(filter(lambda x:x!="", words))
print(result)
```

```python
# Question 12: Use lambda with reduce() to compute factorial.
fact = lambda n: 1 if n==0 else n*fact(n-1)
print(fact(5))

```

```python
# Question 13: Write a lambda that returns larger of two numbers.
large = lambda a,b : a if a>b else b
print(large(10,5))
```

```python
# Question 14: Use lambda to check if number is divisible by 5.
check = lambda x : x%5==0
print(check(25))
```

```python
# Question 15: Use lambda + map() to add 10 to each element.
nums = [1,2,3]
result = list(map(lambda x:x+10, nums))
print(result)
```

```python
# Question 16: Sort list of dictionaries by key "age".
data = [{"name":"Yukta","age":23},{"name":"Parth","age":20}]
result = sorted(data, key=lambda x:x["age"])
print(result)
```

```python
# Question 17: Lambda that returns True if character is vowel.
check = lambda x : x.lower() in "aeiou"
print(check("A"))
```

```python
# Question 18: Extract words of length >5 using filter + lambda.
words = ["python","data","analysis","hi"]
result = list(filter(lambda x:len(x)>5, words))
print(result)
```

```python
# Question 19: Lambda to calculate area of circle.
area = lambda r : 3.14*r*r
print(area(5))
```

```python
# Question 20: Remove duplicates using lambda + set.
nums = [1,2,2,3,4,4]
remove = lambda x : list(set(x))
print(remove(nums))
```

```python
# Question 21: Use lambda with reduce() to find product of all numbers in a list.
nums = [1,2,3,4]
result = 1
for i in nums:
    result = result * i
print(result)
```

```python
# Question 22: Write a lambda that returns absolute value of a number.
abs_val = lambda x : x if x>=0 else -x
print(abs_val(-7))
```

```python
# Question 23: Use lambda to sort a list of strings by their length.
words = ["Yukta","hi","data","yamaha"]
result = sorted(words, key=lambda x:len(x))
print(result)
```

```python
# Question 24: Use lambda to get only uppercase characters from a string.
text = "PyTHonDAta"
result = list(filter(lambda x:x.isupper(), text))
print(result)
```

```python
# Question 25: Return square if number is even, cube if odd.
check = lambda x : x*x if x%2==0 else x*x*x
print(check(4))
print(check(3))
```

```python
# Question 26: Convert Celsius to Fahrenheit using lambda.
convert = lambda c : (c*9/5)+32
print(convert(25))
```

```python
# Question 27: Check if two strings are anagrams.
ana = lambda a,b : sorted(a)==sorted(b)
print(ana("listen","silent"))
```

```python
# Question 28: Extract only numeric values from a mixed list.
data = [1,"yukta",3,"bizmetric",5]
result = list(filter(lambda x:type(x)==int, data))
print(result)
```

```python
# Question 29: Use lambda inside any() to check if any element is negative.
nums = [5,3,-2,8]
result = any(map(lambda x:x<0, nums))
print(result)
```
