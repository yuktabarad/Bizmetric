```python
#ATM style menu:
balance=10000
while True:
    menu=input("1. To check balance" \
    "2. To withdraw amount" \
    "3. To exit")

    if menu=="1":
        print(f"The account balance is {balance}")
        wit=input("Want to withdraw amount?(Yes/No)")
        if wit.lower()=="yes":
            money=int(input("Enter amount you want to withdraw"))
            if money>balance:
                print("Insufficient balance")
            else:
                print(f"After withdrawing {money}, the balance is",balance-money)
            
        elif wit.lower()=="no":
            print("Thank you for visiting!!")
            
    
    elif menu=="2":
        mone=int(input("How much money you want to withdraw?"))
        if mone>balance:
                print("Insufficient balance")
        else:
            print("Money is withdrawen")
            print(f"After withdrawing {mone}, the balance is",balance-mone)
        

    elif menu=="3":
        print("Thank you, Come again!")
        break

    else:
        print("You have entered invalid number")
```
